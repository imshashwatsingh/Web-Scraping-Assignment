```python
import string

def remove_punctuation_from_strings(string_list):
    # Create a translation table that maps punctuation to None
    translator = str.maketrans('', '', string.punctuation)
    
    # Use list comprehension to apply the translation table to each string
    return [s.translate(translator) for s in string_list]

data = remove_punctuation_from_strings(data)

print(len(data))

data_file = open("clean_data_sample.txt","w")
for x in data:
    data_file.write(x+" ")
```

```python
# Extracting Derived variables
# We convert the text into a list of tokens using the nltk tokenize module and 
# use these tokens to calculate the 4 variables described below:

# Positive Score: This score is calculated by assigning the value of +1 
# for each word if found in the Positive Dictionary and then adding up all the values.

# Negative Score: This score is calculated by assigning the value of -1
# for each word if found in the Negative Dictionary and then adding up all the values.
# We multiply the score with -1 so that the score is a positive number.

# Polarity Score: This is the score that determines if a given text is positive or negative in nature. 
# It is calculated by using the formula: 
# Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)
# Range is from -1 to +1

# Subjectivity Score: This is the score that determines if a given text is objective or subjective. 
# It is calculated by using the formula: 
# Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
# Range is from 0 to +1
```

```python
positive_score = 0
p = open("MasterDictionary/positive-words.txt","r")
positive_words = p.read().lower().split()
for x in data:
    if x in positive_words:
        positive_score +=1

positive_score
```

```python
negative_score = 0
p = open("MasterDictionary/negative-words.txt","r")
negative_words = p.read().lower().split()
for x in data:
    if x in negative_words:
        negative_score -=1

negative_score *= -1
negative_score
```

```python
polarity_score =(positive_score - negative_score)/((positive_score + negative_score) + 0.000001)
polarity_score
```

```python
total_words_after_cleaning = len(data)
subjectivity_score = (positive_score+negative_score)/((total_words_after_cleaning)+0.000001)
subjectivity_score
```

```python
# Analysis of Readability
# Analysis of Readability is calculated using the Gunning Fox index formula described below.
# Average Sentence Length = the number of words / the number of sentences
# Percentage of Complex words = the number of complex words / the number of words 
# Fog Index = 0.4 * (Average Sentence Length + Percentage of Complex words)
```

```python
fn = open("articles/blackassign0001.txt","r")

text = fn.read().lower().strip()
fn.seek(0)
text2 = fn.read().strip()

def count_lines(text):
    lines = text.splitlines()
    line_count = 0
    for line in lines:
        # If the line is not empty, count it as a line
        if line.strip():
            line_count += 1
    return line_count


def count_words(text):
    # Split the text into words using whitespace as the delimiter
    words = text.split()
    # Return the number of words
    return len(words)

import re

def count_sentences(text):
    # Regular expression to split text at sentence-ending punctuation marks
    sentence_endings = re.compile(r'[.!?]')
    
    # Split the text into sentences
    sentences = sentence_endings.split(text)
    
    # Filter out empty strings and count the remaining sentences
    sentence_count = sum(1 for sentence in sentences if sentence.strip())
    
    return sentence_count

number_of_lines = count_lines(text)
print(number_of_lines)
number_of_words = count_words(text)
print(number_of_words)
number_of_sentences = count_sentences(text)
print(number_of_sentences)
```

```python
avg_sentence_length = number_of_words/number_of_sentences
avg_sentence_length
```

```python
def count_syllables(word):
    word = word.lower().strip()
    
    # Handle special cases for 'es' and 'ed'
    if word.endswith('es') and len(word) > 2 and word[-3] not in 'aeiouy':
        word = word[:-2]
    elif word.endswith('ed') and len(word) > 2 and word[-3] not in 'aeiouy':
        word = word[:-2]
    
    # Removing non-alphabetic characters for cleaner syllable counting
    word = re.sub(r'[^a-z]', '', word)
    
    # Count the number of vowels in the word
    syllable_count = len(re.findall(r'[aeiouy]', word))
    
    # Ensure at least one syllable for every word
    if syllable_count == 0:
        syllable_count = 1
    
    return syllable_count


def syllables_per_word(string_list):
    syllable_counts = []
    for sentence in string_list:
        words = sentence.split()
        for word in words:
            syllable_counts.append((word, count_syllables(word)))
    return sum(syllable_counts)


def count_complex_words(string_list):
    count = 0
    # complex_words = []
    for sentence in string_list:
        words = sentence.split()
        for word in words:
            syllable_count = count_syllables(word)
            if syllable_count > 2:
                # complex_words.append(word)
                count+=1
    return count
```

```python
complex_word_count = count_complex_words(text)
per_conmplex_words = complex_word_count/number_of_words
fog_index = 0.4*(avg_sentence_length+per_conmplex_words)
avg_number_of_words_per_sentence = number_of_words/number_of_sentences

def sum_of_characters(text):
    count=0
    words = text.split()
    for word in words:
        count+=len(word)
    return count

avg_word_length = sum_of_characters(text)/number_of_words

def count_personal_pronouns(text):
    # Define a regex pattern for personal pronouns
    pronoun_pattern = re.compile(r'\b(I|we|my|ours|us)\b', re.IGNORECASE)
    
    # Find all matches in the text
    matches = pronoun_pattern.findall(text)
    
    # Filter out 'US' that should not be counted
    filtered_matches = [match for match in matches if match.lower() != 'us' or re.search(r'\bUS\b', text) is None]
    
    return len(filtered_matches)

personal_pronouns = count_personal_pronouns(text2)

personal_pronouns
```

```python
def count_syllables(word):
    word = word.lower().strip()
    
    # Handle special cases for 'es' and 'ed'
    if word.endswith('es') and len(word) > 2 and word[-3] not in 'aeiouy':
        word = word[:-2]
    elif word.endswith('ed') and len(word) > 2 and word[-3] not in 'aeiouy':
        word = word[:-2]
    
    # Remove non-alphabetic characters for cleaner syllable counting
    word = re.sub(r'[^a-z]', '', word)
    
    # Count the number of vowels in the word
    syllable_count = len(re.findall(r'[aeiouy]', word))
    
    # Ensure at least one syllable for every word
    if syllable_count == 0:
        syllable_count = 1
    
    return syllable_count

def syllables_in_text(text):
    count = 0
    words = re.findall(r'\b\w+\b', text)
    for word in words:
        count+=  count_syllables(word)
    return count


sameple_text = """This is an example sentence. It includes words like running, tested, and foxes. Additionally, extraordinary measures were taken."""
syllable_counts = syllables_in_text(sameple_text)

syllable_counts
```
